1. Objetivo

O objetivo deste projeto √© demonstrar um fluxo simples de ETL (Extract, Transform, Load) utilizando arquivos CSV.

2. Etapa 1 ‚Äì Extra√ß√£o (Extract)

Nesta etapa, os arquivos CSV contendo as informa√ß√µes de clientes e vendas foram carregados para o ambiente local.

```python
import pandas as pd

df_clientes = pd.read_csv('/home/davi/ProjetoDados/data/raw/clientes.csv')
df_vendas = pd.read_csv('/home/davi/ProjetoDados/data/raw/vendas.csv')
```

3. Etapa 2 ‚Äì Transforma√ß√£o (Transform)

Nesta etapa foram aplicadas transforma√ß√µes para preparar os dados para an√°lise e integra√ß√£o.

üßπ 3.1 Ajuste de tipos de dados

A coluna data_venda foi convertida de object para datetime, garantindo que opera√ß√µes temporais sejam poss√≠veis.

```python
df_vendas['data_venda'] = pd.to_datetime(df_vendas['data_venda'])
df_vendas.info()
```

üîó 3.2 Verifica√ß√£o de integridade

Verificar se todos os id_cliente presentes na tabela de vendas existem na tabela de clientes.

```python
df_vendas['id_cliente'].isin(df_clientes['id_cliente'])
```

üîÄ 3.3 Jun√ß√£o das tabelas

As tabelas de clientes e vendas foram unificadas utilizando merge com base no id_cliente.

```python
df = pd.merge(df_clientes, df_vendas, on='id_cliente', how='left')
```

```python
df.duplicated().sum() 

df.isna().sum()
```

4. Etapa 3 ‚Äì Carga (Load)

Ap√≥s as valida√ß√µes, o dataset consolidado foi salvo em um novo arquivo CSV.
